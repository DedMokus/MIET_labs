```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = complex(256,0)
r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,9)
zroot = r**(1/8)*( np.cos((phi+2.0*np.pi*k)/8) + 1j*np.sin((phi+2.0*np.pi*k)/8))
#print(zroot)
plt.title('$\sqrt[8]{256}$ complex roots')

plt.plot(np.real(zroot), np.imag(zroot), '*g')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('equal')
t = np.linspace(0,2*np.pi,100)
plt.plot(r**(1/8)*np.cos(t),r**(1/8)*np.sin(t),':')
plt.grid(True)
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')

print(*zroot)
print(np.real(zroot)<0)
print(np.isclose(np.real(zroot),0))
print(np.angle(zroot) > np.pi/3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function $y = e^{x}$')
plt.show()
```

```python
x = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y = np.sin(x)
x1 = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y1 = np.sin(x)-2
x2 = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y2 = np.sin(x)+1
fig1 = plt.figure() # графическое окно figure 1
ax1 = fig1.add_subplot(1, 1, 1) 
plt.plot(x,y,label = 'y = sin(x)')
plt.plot(x1,y1,':ro',label = 'y = sin(x)-2')
plt.plot(x2,y2,'-.bd',label = 'y = sin(x)+1')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
#plt.legend()
plt.grid(True)
```

Упражнение 3

```python
x = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y = np.cos(x)
x1 = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y1 = np.cos(x)*2
x2 = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y2 = np.cos(x)*0.3
x3 = np.linspace(-2.0*np.pi,2.0*np.pi,100)
y3 = np.cos(x)*(-1)
fig1 = plt.figure() # графическое окно figure 1
ax1 = fig1.add_subplot(1, 1, 1) 
plt.plot(x,y,label = 'y = cos(x)')
plt.plot(x1,y1,':ro',label = 'y = 2cos(x)')
plt.plot(x2,y2,'-.bd',label = 'y = 0.3cos(x)')
plt.plot(x3,y3,'--+k',label = 'y = -cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.grid(True)
```

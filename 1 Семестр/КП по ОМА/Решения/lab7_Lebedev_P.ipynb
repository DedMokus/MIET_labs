```python
x = sp.Symbol('x')

f31 = (3*x)**np.e - np.e**(3*x)
r31 = sp.nsolve(f31,0)
print('x = ',r31)
print('f(x) = ', r31.subs(x,r31))

f32 = x/3 - sp.sin(x)
r32 = sp.nsolve(f32,0)
print('x = ',r32)
print('f(x) = ', r32.subs(x,r32))

```

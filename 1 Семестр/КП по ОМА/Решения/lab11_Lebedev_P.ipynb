```python
def f(x):
    return np.e**(-x)*np.sin(3*np.pi*x)

def der(x_p):
    x = sp.symbols('x')
    y = -1*sp.E**(-1*x)*sp.sin(3*sp.pi*x) + 3*sp.pi*sp.E**(-1*x)*sp.cos(3*sp.pi*x)
    return y.subs(x,x_p)

x = np.linspace(0,2,1000)
plt.plot(x,np.e**(-x)*np.sin(3*np.pi*x),'r')
plt.grid()
minimums = []
maximums = []
nulls=[]
for i in np.arange(0,2,0.1):
    null = round(float(optimize.broyden1(der,i)),3)
    if (null>=0 and null<=2) and (null not in nulls):
        plt.plot(null,f(null),'or')
        nulls.append(null)
    minim = optimize.fmin(f,i,full_output=True,disp=False)
    minim = round(float(minim[0]),3)
    
for null in nulls:
    if f(null)>0:
        maximums.append(null)
    else:
        minimums.append(null)
        
print(maximums)
print(minimums)
```

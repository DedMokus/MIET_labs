```python
a,b,c = symbols('a b c')
p = (['a','-b','c'])
q = ([0,'b','a'])
r = ([0,'b','-c'])
s = (['a','b','c'])
A = Matrix([p,q,r])
if (det(A)) == 0:
    print('Линейно независимые')
A_det = det(A)
X_m = Matrix(A)
X_m[:, 0] = s
Y_m = Matrix(A)
Y_m[:, 1] = s
Z_m = Matrix(A)
Z_m[:, 2] = s
x = det(X_m) / A_det
y = det(Y_m) / A_det
z = det(Z_m) / A_det
print(x)
print(y)
print(z)
```

Упражнение 3.4

```python
A = np.array([-2,0])
B = np.array([3,4])
C = np.array([4,-1])

AB = B - A
BC = C - B 
CA = A - C


cos_angle1 = np.dot(AB, BC) / np.linalg.norm(AB) / np.linalg.norm(BC)
cos_angle2 = np.dot(BC, CA) / np.linalg.norm(BC) / np.linalg.norm(CA)
cos_angle3 = np.dot(CA, AB) / np.linalg.norm(CA) / np.linalg.norm(AB)

print ('Косинус угла между ab и bc:', cos_angle1)
print ('Косинус угла между bc и ca:', cos_angle2)
print ('Косинус угла между ca и ab:', cos_angle3)
alfa1= np.arccos(abs(cos_angle1)) # арккосинус угла
alfa1=math.degrees(alfa1)
print ('Угол в градусах:', alfa1)

alfa2= np.arccos(abs(cos_angle2)) # арккосинус угла
alfa2=math.degrees(alfa2)
print ('Угол в градусах:', alfa2)

alfa3= np.arccos(abs(cos_angle3)) # арккосинус угла
alfa3=math.degrees(alfa3)
print ('Угол в градусах:', alfa3)
```

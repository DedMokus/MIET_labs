```python
# параметры плоскости
p541 = Plane((1,-1,3),(2,-3,5))
A = 2; B = -3; C = 5; D = -20
f = lambda x, y: (-A * x - B * y - D) / C # строим плоскость
fig = plt.figure(figsize=(10, 10)) # создаём холст
ax = fig.add_subplot(1, 1, 1, projection='3d') # 
xval = np.linspace(-4, 4, 100) # разбиение для x
yval = np.linspace(-4, 4, 100) # разбиение для y
x, y = np.meshgrid(xval, yval)
p = np.array([0, 0, - 3]) # точка на плоскости
v = np.array([3, 4, -4]) # вектор нормали
ax.quiver(*p, *v, color='Blue') # строим вектор нормали
z = f(x, y) # вычисляем значения z в точках
surf = ax.plot_surface(x, y, z, rstride=2, cstride=2, cmap=cm.rainbow) # строим плоскость
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость 2x - 3y + 5z – 20 = 0')
ax.view_init(10, -30)
plt.show()
```

```python
def distpr(l1,l2):
    A = l1.p1
    P = Plane(A,(l1.direction))
    Pl2 = P.projection_line(l2)
    return Pl2.distance(A)

linec1 = Line(Point(2,-1,0),Point(5,3,2))
linec2 = Line(Point(7,1,3),Point(10,5,5))

distpr(linec1,linec2)
    
```

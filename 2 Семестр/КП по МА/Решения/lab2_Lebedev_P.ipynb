```python
import numpy as np
import sympy as sp
import scipy as sc
import matplotlib.pyplot as plt
```

Упражнение 1

```python
x = np.linspace(0,np.pi/2,1000)
y = x*np.cos(x)
plt.plot(x,y,'r')
plt.grid()


x = sp.Symbol('x')
sp.pprint(sp.integrate(x*sp.cos(x),(x,0,sp.pi/2)))
```

Упражнение 3

```python
t = np.linspace(0,2*np.pi,1000)
x = np.cos(t)
y = np.sin(t)
plt.plot(x,y,'k')
plt.axvline(0)
plt.axhline(0)
plt.grid()

t = sp.Symbol('t')
x = sp.cos(t)
y = sp.sin(t)
dx = sp.diff(x)
t1 = sp.cos(0)
t2 = sp.cos(sp.pi/2)

S = 4*sp.integrate(y*dx,(t,t1,t2))
print(S)
```

```python
t = np.linspace(0,2*np.pi,1000)
x = 2*np.cos(t)
y = 3*np.sin(t)
plt.plot(x,y,'k')
plt.axvline(0)
plt.axhline(0)
plt.grid()

t = sp.Symbol('t')
x = 2*sp.cos(t)
y = 3*sp.sin(t)
dx = sp.diff(x)
t1 = sp.cos(0)
t2 = sp.cos(sp.pi/2)

S = 4*sp.integrate(y*dx,(t,t1,t2))
print(S)
```

```python
t = np.linspace(0,2*np.pi,1000)
r = np.exp(t)
x = r*np.cos(t)
y = r*np.sin(t)
plt.subplot(2,1,1)
plt.plot(x,y,'k')

plt.axvline(0)
plt.axhline(0)
plt.grid()
plt.subplot(2,1,2,projection='polar')
plt.polar(x,y,'r')

t = sp.Symbol('t')
r = sp.exp(t)
x = r*sp.cos(t)
y = r*sp.sin(t)
t1 = sp.cos(0)
t2 = sp.cos(sp.pi*2)

S = 1/2*sp.integrate(r**2,(t,t1,t2))
print(S)
```

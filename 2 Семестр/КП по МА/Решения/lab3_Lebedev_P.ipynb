1. Создайте функцию, которая строит график последовательности частичных сумм ряда. В качестве входных параметров функции использовать
формулу
$a_n$
общего члена последовательности и число
$n_0$
рассматриваемых членов

```python
def summ_plot(an, n0):
    x = np.arange(1, 1+n0)
    y = np.array([])
    cs = 0
    for i in range(len(x)):
        cs += an(x[i])
        y = np.append(y, cs)
    plt.scatter(x, y)
    plt.grid()
    plt.show()

n = sp.Symbol('n')
func =lambda n: 3**n
summ_plot(func,10)
```

2. а) Используя определение, установить сходимость иди расходимость рядов
$\sum_{n=1}^\infty q^\alpha$
для нескольких значений
$0 < q < 1, q = 1, q > 1$. В случае сходимости ряда
найти его сумму.

```python
def is_shod(func):
    
    Sum = 0
    eps = 1e-4
    i = 1
    for i in range(1,1000):
        a = func(i)
        #print(a,'  ',Sum)
        Sum += a
        
    if (-10 < round(Sum) < 10):
        print('Сумма ряда = ',round(Sum,10))
    elif (round(Sum) < -10):
        print('Сумма ряда = ',-sp.oo)
    elif (round(Sum) > 10):
        print('Сумма ряда = ',sp.oo)
    
    if (Sum > 10 or Sum < -10 or Sum == 0):
        return False
    else:
        return True
    
qs = [0.5,1,1.5]
for q in qs:
    print(f'q = {q}')
    n = sp.Symbol('n')
    func = lambda n: q**n
    print(is_shod(func))
```

```python
def sumtoch(f,e):
    s = 0
    n0 = 1
    q = sp.limit(abs(f(n+1)) / abs(f(n)), n, sp.oo)
    while abs(f(n0+1)) / (1-q) > e:
        s += f(n0)
        n0 += 1
    return n0, sp.N(s)
```

```python
f = lambda n: 1/(n+1)**n
print(sumtoch(f, 0.0001))
```

```python
f = lambda n: (-1)**(n-1)/sp.factorial(n)
print(sumtoch(f, 0.0001))
```
